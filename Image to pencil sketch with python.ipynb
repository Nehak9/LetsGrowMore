# Task_4:- Image to Pencil Sketch With Python

# importing libraries

```python
import numpy as np
import cv2
from IPython.display import display
from PIL import Image
```

# Read the image in RBG Format

```python
img = cv2.imread('Shahid kapoor image.jpg')
img = cv2.resize(img,(250,300))
RBG_img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display(Image.fromarray(RBG_img))
```

# RBG Image into Greayscale image(White and Black Photo)

```python
grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_img))
```

# Inverted Greayscale Image(Negative Image)

```python
inverted_grey_img=255-grey_img
display(Image.fromarray(inverted_grey_img))
```

# Convert into Negative Image

```python
Negative_img=cv2.bitwise_not(grey_img)
display(Image.fromarray(grey_img))
```

# Smoothen the image

```python
Smooth_img=cv2.GaussianBlur(Negative_img,(5,5),0,0)
display(Image.fromarray(Smooth_img))
```

# Convert into pencil sketch

```python
invert_blur=255-Smooth_img
pencil_Sketch = cv2.divide(grey_img,invert_blur,scale=250.0)
display(Image.fromarray(pencil_Sketch))
```

# Combination of pencil sketch and Image

```python
display(Image.fromarray(RBG_img))
display(Image.fromarray(pencil_Sketch))
```
